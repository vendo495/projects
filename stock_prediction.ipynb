```python
del nifty_next_50["Dividends"]
del nifty_next_50["Stock Splits"]
```

```python
nifty_next_50["Tomorrow"] = nifty_next_50["Close"].shift(-1)
```

```python
nifty_next_50
```

```python
nifty_next_50["Target"] = (nifty_next_50["Tomorrow"] > nifty_next_50["Close"]).astype(int)
```

```python
nifty_next_50
```

```python
nifty_next_50 = nifty_next_50.loc["2016-01-01":].copy()
nifty_next_50
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = nifty_next_50.iloc[:-200]
test = nifty_next_50.iloc[-200:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
def predict(train, test, predictors, model):
  model.fit(train[predictors], train['Target'])
  preds = model.predict(test[predictors])
  preds = pd.Series(preds, index = test.index, name = "Predictions")
  combined = pd.concat([test["Target"], preds], axis = 1)
  return combined
```

```python
def backtest(data, model, predictors, start=738, step=246):
    all_predictions = []

    # print(f"Data length: {data.shape[0]}, Start: {start}, Step: {step}")  # Debug print

    for i in range(start, data.shape[0], step):
        # print(f"Processing slice: {i}:{i+step}")  # Debug print
        train = data.iloc[i:(i+step)].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        # print(f"Predictions shape: {predictions.shape}")  # Debug print
        all_predictions.append(predictions)

    # if not all_predictions:  # Check if list is empty
    #     print("No predictions were made.")
    #     return pd.DataFrame()  # Return empty DataFrame or handle appropriately

    return pd.concat(all_predictions)
```

```python
# print(len(nifty_next_50))
```

```python
predictions = backtest(nifty_next_50, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```

```python
horizons = [2,5,6,246,984]
new_predictors = []

for horizon in horizons:
  rolling_averages = nifty_next_50.rolling(horizon).mean()

  ratio_column = f"Close_Ratio_{horizon}"
  nifty_next_50[ratio_column] = nifty_next_50["Close"] / rolling_averages["Close"]

  trend_column = f"Trend_{horizon}"
  nifty_next_50[trend_column] = nifty_next_50['Target'].rolling(horizon).sum()

  new_predictors+= [ratio_column, trend_column]
```

```python
nifty_next_50 = nifty_next_50.dropna(subset=nifty_next_50.columns[nifty_next_50.columns != "Tomorrow"])
```

```python
nifty_next_50
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split= 10, random_state=1)
```

```python
def predict(train, test, predictors, model):
  model.fit(train[predictors], train['Target'])
  preds = model.predict(test[predictors])
  preds[preds>=.6] = 1
  preds[preds< .6] = 0
  preds = pd.Series(preds, index = test.index, name = "Predictions")
  combined = pd.concat([test["Target"], preds], axis = 1)
  return combined
```

```python
predictions = backtest(nifty_next_50, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
